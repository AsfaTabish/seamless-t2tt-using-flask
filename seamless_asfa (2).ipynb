```python
from flask import Flask, render_template, request
from pyngrok import ngrok
import torch
from seamless_communication.models.inference import Translator

app = Flask(__name__)

# Load your models here
translator_medium = Translator("seamlessM4T_medium", "vocoder_36langs", torch.device("cuda:0"), torch.float16)

@app.route("/", methods=["GET", "POST"])
def index():
    if request.method == "POST":
        message_to_translate = request.form.get("input_text")

        # Translate the input text using the medium model
        translated_text, _, _ = translator_medium.predict(message_to_translate, "t2tt", 'urd', src_lang='eng')

        return render_template("index.html", translation=translated_text, input_text=message_to_translate)

    return render_template("index.html")

# Set the template folder path
app.template_folder = '/content/drive/MyDrive/Colab Notebooks/seamlessM4T/template'

# Get the public URL for the Flask app using ngrok
public_url = ngrok.connect(port=5000)
print("Ngrok URL:", public_url)

# Run the Flask app
if __name__ == "__main__":
    app.run()
```
